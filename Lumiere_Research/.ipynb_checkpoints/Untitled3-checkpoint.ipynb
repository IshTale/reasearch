```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
import re
```

```python
pd.set_option('display.max_colwidth',500)
pd.set_option('display.max_rows',500)
```

```python
play = pd.read_csv("play_by_play.csv")
```

```python
play = play.drop(['eventnum', 'eventmsgactiontype', 'period',
       'wctimestring', 'pctimestring', 'neutraldescription', 'person1type',
       'player1_team_id', 'player1_team_city',
       'player1_team_nickname', 'player1_team_abbreviation', 'person2type',
       'player2_team_id', 'player2_team_city',
       'player2_team_nickname', 'player2_team_abbreviation', 'person3type',
       'player3_id','player3_name','player3_team_id', 'player3_team_city',
       'player3_team_nickname', 'player3_team_abbreviation',
       'video_available_flag'], axis = 1)
play
```

```python
player = pd.DataFrame(data = {'person_id' : [], 'points' : []})
player
```

```python
count = 0
game_id = 0
game_id_index = 0
eventmsgtype = 1
scoremargin = 5
player1_id = 6
margin = 0
print("starting...")
for row in play.values:
    data = {'person_id' : 0, 'points' : 0}
    if (
        (row[eventmsgtype] == 8) | (row[eventmsgtype] == 9) | 
        (row[eventmsgtype] == 10) | (row[eventmsgtype] == 11) | 
        (row[eventmsgtype] == 12) | (row[eventmsgtype] == 13) |
        (row[eventmsgtype] == 2) | (row[eventmsgtype] == 4) |
        (row[eventmsgtype] == 5) | (row[eventmsgtype] == 6) |
        (row[eventmsgtype] == 7) | (row[eventmsgtype] == 3) 
        ):
        continue
    
    elif pd.notna(row[scoremargin]):
        if row[scoremargin] == "TIE":
            points = abs(0 - margin)
        else:
            points = abs(int(row[scoremargin]) - margin)
        if not player.loc[(player["person_id"] == row[player1_id]), "person_id"].empty:
            player.loc[(player["person_id"] == row[player1_id]), "points"] += points
        else:
            data["person_id"] = row[player1_id]
            data["person_id"] = row[player1_id]
            player.loc[(player["person_id"] == row[player1_id]), "points"] = points
            player = player.append(data, ignore_index=True)
    
    if game_id != row[game_id_index]:
        margin = 0
        game_id = row[game_id_index]
    
    if count % 100000 == 0:
        print(count)
    count += 1
print("Finished")
```
